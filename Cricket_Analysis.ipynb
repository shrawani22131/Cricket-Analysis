```python
import json
import os

# Path to the specific JSON file
file_name = "t20_wc_match_results.json"
file_path = os.path.join(inner_folder_path, file_name)  # Correct path to the file

# Print the full file path to verify it's correct
print(f"Attempting to open file at: {file_path}")

# Open and read the JSON file
try:
    with open(file_path, 'r') as file:
        data = json.load(file)  # Load the JSON data into a Python dictionary
    # Print the contents of the file
    print(f"Contents of {file_name}:")
    print(data)
except FileNotFoundError:
    print(f"File not found: {file_path}")
```

```python
df = pd.DataFrame(data[0]['matchSummary'])  
df.head()
```

```python
df.rename({'scorecard':'match_id'}, axis = 1, inplace = True) # change the name scorecard to match_id
df.head()
```

```python
match_ids_dict = {}
for index, row in df.iterrows():
    key1 = row['team1'] + 'Vs' + row['team2']
    key2 = row['team2'] + 'Vs' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
    match_ids_dict

```

```python
match_ids_dict.update({
    'Namibia Vs Sri Lanka': 'T201 # 1823',
    'Sri Lanka Vs Netherlanda': 'T201 # 1823',
    'U.A.E Vs Netherlanda': 'T201 # 1825',
    'Netherlanda Vs U.A.E': 'T201 # 1825'
})
```

# Batting Summary

```python
# Path to the specific JSON file
file_name = "t20_wc_batting_summary.json"
file_path = os.path.join(inner_folder_path, file_name)  # Correct path to the file

# Print the full file path to verify it's correct
print(f"Attempting to open file at: {file_path}")

# Open and read the JSON file
try:
    with open(file_path, 'r') as file:
        data = json.load(file)  # Load the JSON data into a Python dictionary
    # Print the contents of the file
    print(f"Contents of {file_name}:")
    print(data)
except FileNotFoundError:
    print(f"File not found: {file_path}")
```

```python
all_records = []
for rec in data:
    all_records.extend(rec['battingSummary'])
    
df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
df_batting['out/not_out']= df_batting.dismissal.apply(lambda x: 'out' if len (x) > 0 else 'not_out') # new col out/not_out
df_batting.head(11)
```

```python
df_batting.drop(columns=['dismissal'], inplace = True) # remove colums 'dismissal'
df_batting.head()
```

```python
df_batting.head(11)
```

```python
df_batting['batsmanName']= df_batting ['batsmanName'].apply(lambda x:x.replace('â€',' ')) # remove the symbol
df_batting['batsmanName']= df_batting ['batsmanName'].apply(lambda x:x.replace('\xa0',' '))  
df_batting.head(11)                                                     
```

```python
df_batting['match_id'] = df_batting['match'] .map( match_ids_dict )
df_batting.head()
```
